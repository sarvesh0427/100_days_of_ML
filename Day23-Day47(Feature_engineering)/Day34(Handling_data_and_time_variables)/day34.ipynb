# Working with dates

```python
# converting date to datetime datatype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

### 1.Extract the year

```python
date['date_year'] = date['date'].dt.year
```

```python
date.head()
```

### 2.Extract the month

```python
date['date_month'] = date['date'].dt.month
```

```python
date['date_month_name'] = date['date'].dt.month_name()
```

```python
date.head()
```

### 3. Extract the days

```python
date['date_day'] = date['date'].dt.day
```

```python
# day name
date['date_day_name'] = date['date'].dt.day_name()
```

```python
date.head()
```

```python
# day of week 
date['date_day_dow'] = date['date'].dt.dayofweek
```

```python
date.head()
```

```python
# is weekend ?
date['date_weekend'] = np.where(date['date_day_name'].isin(['Sunday','Saturday']),1,0)
date = date.drop(columns=['product_id','city_id','orders'])
```
