# Gradient descent step by step 

```python
import numpy as np
from sklearn.datasets import make_regression
```

```python
X,y = make_regression(n_samples=4,n_features=1, n_informative=1, n_targets=1, noise =80, random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

### Ordinary least squared method

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
# before prediction
plt.scatter(X,y)
plt.plot(X,y)

# after prediction
plt.plot(X,reg.predict(X), color='red')
plt.scatter(X,reg.predict(X), color='blue')
```

### Gradient descent method

```python
# assuming slope is constant = 78.35
# lets assume the starting value for intercept b =0

y_pred =((78.35*X)+0).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X), color='red', label ='OLS')
plt.plot(X,y_pred, color='green', label='b=0')
plt.legend()
plt.show()
```

```python
m = 78.35
b = 0
```

```python
# iteration 1

loss_slope = -2*np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
# learning rate = 0.1
lr = 0.1
step_size = loss_slope * lr
step_size
```

```python
# calculate new intercept
b = b- step_size
b
```

```python
y_pred1 = ((78.35*X)+b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='black',label='OLS')
plt.plot(X, y_pred, color='blue', label='b=0')
plt.plot(X,y_pred1, color='blue', alpha = 0.6, label='b={}'.format(b))
plt.legend()
```

```python
# iteration 2

loss_slope = -2*np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
step_size= lr*loss_slope
```

```python
b = b- step_size
```

```python
y_pred2 = ((78.35*X)+b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='black',label='OLS')
plt.plot(X, y_pred, color='blue', label='b=0')
plt.plot(X,y_pred1, color='blue', alpha = 0.6, label='b= 1st iteration')
plt.plot(X,y_pred2, color='red', alpha=0.8, label='b= 2nd iteration')
plt.legend()
```

```python
# Iteration 3

loss_slope = -2 *np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b- step_size
```

```python
y_pred3 = ((78.35*X)+b).reshape(4)

plt.figure(figsize=(15,15))
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='black',label='OLS')
plt.plot(X, y_pred, color='blue', label='b=0')
plt.plot(X,y_pred1, color='blue', alpha = 0.6, label='b= 1st iteration')
plt.plot(X,y_pred2, color='red', label='b= 2nd iteration')
plt.plot(X,y_pred3, color='green', alpha =0.8, label='b= 3rd iteration')
plt.legend()
```

### Using loop for entire process

```python
b = -100
m = 78.35
lr =0.1

epochs = 10

for i in range(epochs):
    loss_slope = -2* np.sum(y-m*X.ravel() -b)
    step_size = lr* loss_slope
    b = b - step_size
    y_pred = m*X+b
    plt.plot(X,y_pred)

plt.scatter(X,y)
```
