# DBSCAN Clustering Algorithms | Density Based Clustering

```python
# importing all required libraries

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import DBSCAN
```

```python
# creating dataset

X = np.array([[1, 2], [2, 2], [2, 3],[8, 7], [8, 8], [25, 80]])
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
db = DBSCAN(eps=10, min_samples=2)

db.fit(X)
db.labels_
```

```python
# Create a concentric circle dataset

X, _ = make_circles(n_samples=500, factor=.5, noise=.03, random_state=4)
```

```python
# Apply DBSCAN to the dataset

dbscan = DBSCAN(eps=0.1, min_samples=5)
clusters = dbscan.fit_predict(X)
```

```python
# visualizing the clusters

plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='viridis', marker='o')
plt.title("DBSCAN Clustering of Concentric Circles")
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.show()
```

### Limitation of DBSCAN

- Very sensitive to hyperparameters

```python
dbscan = DBSCAN(eps=0.4, min_samples=5)  # changes to hyperparameter
clusters = dbscan.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='viridis', marker='o')
plt.title("DBSCAN Clustering of Concentric Circles")
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.show()
```
